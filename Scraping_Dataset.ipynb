```python
import pandas as pd
import csv
from google_play_scraper import app, reviews, Sort, reviews_all
```

# Scraping Dataset

Pada proyek ini, saya akan melakukan analisis sentimen terhadap ulasan dan pendapat pengguna terkait aplikasi "Spotify: Music and Podcasts" di Play Store.

```python
scrapreview,_ = reviews(
    'com.tokopedia.tkpd',   # ID aplikasi
    lang='id',             # Bahasa ulasan (default: 'en')
    country='id',          # Negara (default: 'us')
    sort=Sort.MOST_RELEVANT, # Urutan ulasan (default: Sort.MOST_RELEVANT)
    count=12000             # Jumlah maksimum ulasan yang ingin diambil
)
```

```python
# Menyimpan ulasan dalam file CSV
df = pd.DataFrame(scrapreview)

df.to_csv("ulasan_aplikasi_tokped.csv", index=False)
```
